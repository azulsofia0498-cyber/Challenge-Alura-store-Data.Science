### Importación de datos



```python
import pandas as pd

# URLs de los CSV originales
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargamos los DataFrames correctamente
tienda = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda ["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print("Ingresos por tienda:")
print(f"Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")
```

```python
print(tienda.columns.tolist())
```

# 2. Ventas por categoría

```python
# Contar productos vendidos por categoría usando la columna correcta
cat_tienda1 = tienda["Categoría del Producto"].value_counts()
cat_tienda2 = tienda2["Categoría del Producto"].value_counts()
cat_tienda3 = tienda3["Categoría del Producto"].value_counts()
cat_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Mostrar resultados
print("\nProductos vendidos de cada tienda y su categoría:")

print("\nTienda 1")
print(cat_tienda1)

print("\nTienda 2")
print(cat_tienda2)

print("\nTienda 3")
print(cat_tienda3)

print("\nTienda 4")
print(cat_tienda4)
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

print("Calificaciones promedio por tienda:")
print(f"Tienda 1: {promedio_tienda1:.2f}")
print(f"Tienda 2: {promedio_tienda2:.2f}")
print(f"Tienda 3: {promedio_tienda3:.2f}")
print(f"Tienda 4: {promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Tienda 1
ventas1 = tienda["Producto"].value_counts()
mas_vendido1 = ventas1.idxmax()
veces_mas1 = ventas1.max()
menos_vendido1 = ventas1.idxmin()
veces_menos1 = ventas1.min()

# Tienda 2
ventas2 = tienda2["Producto"].value_counts()
mas_vendido2 = ventas2.idxmax()
veces_mas2 = ventas2.max()
menos_vendido2 = ventas2.idxmin()
veces_menos2 = ventas2.min()

# Tienda 3
ventas3 = tienda3["Producto"].value_counts()
mas_vendido3 = ventas3.idxmax()
veces_mas3 = ventas3.max()
menos_vendido3 = ventas3.idxmin()
veces_menos3 = ventas3.min()

# Tienda 4
ventas4 = tienda4["Producto"].value_counts()
mas_vendido4 = ventas4.idxmax()
veces_mas4 = ventas4.max()
menos_vendido4 = ventas4.idxmin()
veces_menos4 = ventas4.min()

# Mostrar resultados
print("Productos más y menos vendidos por tienda:\n")
print(f"Tienda 1 → Más vendido: {mas_vendido1} ({veces_mas1} veces), Menos vendido: {menos_vendido1} ({veces_menos1} veces)")
print(f"Tienda 2 → Más vendido: {mas_vendido2} ({veces_mas2} veces), Menos vendido: {menos_vendido2} ({veces_menos2} veces)")
print(f"Tienda 3 → Más vendido: {mas_vendido3} ({veces_mas3} veces), Menos vendido: {menos_vendido3} ({veces_menos3} veces)")
print(f"Tienda 4 → Más vendido: {mas_vendido4} ({veces_mas4} veces), Menos vendido: {menos_vendido4} ({veces_menos4} veces)")
```

# 5. Envío promedio por tienda

```python
# Tienda 1
costo_envio1 = tienda["Costo de envío"].mean()

# Tienda 2
costo_envio2 = tienda2["Costo de envío"].mean()

# Tienda 3
costo_envio3 = tienda3["Costo de envío"].mean()

# Tienda 4
costo_envio4 = tienda4["Costo de envío"].mean()

# Mostrar resultados
print("Costo de envío promedio por tienda:")
print(f"Tienda 1: ${costo_envio1:.2f}")
print(f"Tienda 2: ${costo_envio2:.2f}")
print(f"Tienda 3: ${costo_envio3:.2f}")
print(f"Tienda 4: ${costo_envio4:.2f}")
```

```python
tiendas_df = [tienda, tienda2, tienda3, tienda4]
nombres = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

import matplotlib.pyplot as plt

for df, nombre in zip(tiendas_df, nombres):
    categorias = df["Categoría del Producto"].value_counts()
    plt.figure(figsize=(7,7))
    plt.pie(categorias, labels=categorias.index, autopct="%1.1f%%", startangle=90, colors=plt.cm.Pastel1.colors)
    plt.title(f"Distribución de categorías de productos - {nombre}", fontsize=14)
    plt.show()
```

```python
for df, nombre in zip(tiendas_df, nombres):
    top_productos = df["Producto"].value_counts().head(5)
    plt.figure(figsize=(8,5))
    plt.barh(top_productos.index, top_productos.values, color='lightgreen')
    plt.title(f"Top 5 productos más vendidos - {nombre}", fontsize=14)
    plt.xlabel("Cantidad vendida")
    plt.ylabel("Producto")
    plt.gca().invert_yaxis()  # Para que el más vendido quede arriba
    plt.show()
```

Tras analizar ingresos, satisfacción de clientes, variedad de productos y costos de envio, la Tienda 1 resulta la mejor opción para el señor Juan, ya que combina buen rendimiento y equilibrio en sus indicadores.En cambio, la Tienda 4 muestra bajo desempeño en venta, menor satisfacción del cliente y poca rotación de productos, por lo que no sería recomendable.
